```python
def swap_values(x, y):
    # Check if both x and y are numeric 
    if not (isinstance(x, (int, float)) and isinstance(y, (int, float))):
        return -1
   
    # Swap using arithmetic operations (only x and y as variables)
    x = x + y
    y = x - y
    x = x - y

    print(f"x = {x}, y = {y}")
    
 
    
```

```python
swap_values("Apple", 10)
```

```python
swap_values(9,17)
```
