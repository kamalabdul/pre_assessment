```python
def string_reverse(s):
    if not isinstance(s, str):
        raise TypeError("s must be a string.")

    reversed_str = ""
    i = len(s) - 1
    while i >= 0:
        reversed_str += s[i]
        i -= 1
    return reversed_str
    
```

```python
string_reverse ("Hello World")
```

```python
string_reverse ("Python")
```

```python
string_reverse (4444444)
```
